## What is IPython?
IPython is a project to improve python's REPL. 
Headed by Fernando Perez
### Brief History
* Started in 2001
* Backend & Frontend seperated in 2009
* Notebook 2011

## What is IPython notebook?
The Notebook is an html interface to the ipython kernel over zeromq.
## What is the IRuby notebook?
IPython's notebook html interface and controller with a ruby kernel running pry or irb.
Currently maintained by Daniel Mendler
